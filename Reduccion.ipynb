```python
# Diccionario de categorías
categorias = {
    1: ["NOMINA TRADICIONAL", "NOMINA TRADICIONAL REMODELA", "NOMINA TRADICIONAL MEDIOS DIGITALES"],
    2: ["DOMICILIACION TRADICIONAL", "DOMICILIACION REMODELA", "DOMICILIACIÓN EXPRESS"],
    3: ["REFINANCIAMIENTO NOMINA TRADICIONAL", "REFINANCIAMIENTO MEDIOS ELECTRONICOS NOMINA TRADICIONAL", 
        "REFINANCIAMIENTO INMEDIATO NOMINA TRADICIONAL", "REFINANCIAMIENTO SALDO INSOLUTO NOMINA TRADICIONAL", 
        "REFINANCIAMIENTO CONSTANTE NOMINA TRADICIONAL", "REFINANCIAMIENTO DOMICILIACION TRADICIONAL", 
        "REFINANCIAMIENTO PERSONALIZADO", "REFINANCIAMIENTO NOMINA TRADICIONAL MEDIOS DIGITALES"],
    4: ["VENTA DE CARTERA NOMINA TRADICIONAL", "VENTA DE CARTERA NOMINA TADICIONAL"],
    5: ["RENOVACION NOMINA TRADICIONAL", "NOM COMPRA DE CARTERA BARACAF"],
    6: ["LIQUIDACION A COMPETIDORES NOMINA TRADICIONAL", "BROXEL"]
}

# Crear una función para asignar categorías
def asignar_categoria(producto):
    for categoria, productos in categorias.items():
        if producto in productos:
            return categoria
    return None  # Si no coincide con ninguna categoría

# Aplicar la función al DataFrame y agregar la nueva columna 'Categoria' a la derecha
df['Prod_cat'] = df['PRODUCTO'].apply(asignar_categoria)

# Mostrar el DataFrame con la nueva columna agregada
df.head(5)
```

```python
# Función para asignar categoría numérica 1 o 2 basado en el nombre del producto
def asignar_categoria_numerica(producto):
    if 'DIGITAL' in producto:
        return 2  # DIGITAL
    else:
        return 1  # TRADICIONAL

# Aplicar la función al DataFrame y agregar la nueva columna 'Categoria' a la derecha
df['Colo_cat'] = df['COLOCACION'].apply(asignar_categoria_numerica)
df.head(5)
```

```python
# Función para asignar categoría numérica al tipo de pago
def asignar_pago_categoria(tipo_pago):
    if tipo_pago == 'QUINCENAL':
        return 1
    elif tipo_pago == 'MENSUAL':
        return 2
    elif tipo_pago == 'CATORCENAL':
        return 3
    elif tipo_pago == 'DIARIO':
        return 4
    else:
        return None  # Si no hay coincidencia

# Aplicar la función para crear la columna 'Categoria_Pago'
df['FrecPago_cat'] = df['FREC_PAGO'].apply(asignar_pago_categoria)
df.head(5)
```

```python
# Función para asignar categoría numérica al género
def asignar_genero_categoria(genero):
    if genero == 'FEMENINO':
        return 1
    elif genero == 'MASCULINO':
        return 2
    else:
        return None  # Si no hay coincidencia

# Aplicar la función para crear la columna 'Categoria_Genero'
df['Gen_Cat'] = df['SEXO'].apply(asignar_genero_categoria)
df.head(5)
```

```python
# Función para asignar categoría numérica al canal
def asignar_canal_categoria(canal):
    if pd.isna(canal):  # Si el valor es NaN
        return 0
    elif '(LLS) Llamada de Salida' in canal:
        return 1
    elif '(WTS) Whats App' in canal:
        return 2
    elif '(SMS) SMS' in canal:
        return 3
    elif '(LLE) Llamada de Entrada' in canal:
        return 4
    elif '(EML) E-Mail' in canal:
        return 5
    elif '(JUD) Judicial' in canal:
        return 6
    else:
        return 0  # Si el valor no coincide con ninguno

# Aplicar la función para crear la columna 'Categoria_Canal'
df['CanalCat'] = df['canal_final'].apply(asignar_canal_categoria)
df.head(5)
```

```python
# Función para asignar categoría numérica al registro
def asignar_registro_categoria(registro):
    if pd.isna(registro):  # Si el valor es NaN
        return 0
    elif '(NEQ) Numero Equivocado' in registro:
        return 1
    elif '(MCF) Mensaje con Familiar' in registro:
        return 2
    elif '(MCR) Mensaje con Referencia' in registro:
        return 3
    elif '(LTP) Liquidación Total en Pagos' in registro:
        return 4
    elif '(NDP) Negativa de Pago' in registro:
        return 5
    elif '(EXI) Exitoso' in registro:
        return 6
    elif '(PDP) Promesa de Pago' in registro:
        return 7
    elif '(CCT) Contacto con Titular' in registro:
        return 8
    elif '(PPV) Promesa Vigente' in registro:
        return 9
    elif '(PPP) Promesa de Pago Parcial' in registro:
        return 10
    elif '(LQP) Liquidación con Quita en Pagos' in registro:
        return 11
    elif '(ACL) Aclaracion' in registro:
        return 12
    elif '(NEG) En Negociacion' in registro:
        return 13
    elif '(LQU) Liquidación con Quita en Una Exhibición' in registro:
        return 14
    elif '(LLM) Llamar mas Tarde' in registro:
        return 15
    elif '(PPR) Promesa Rota' in registro:
        return 16
    elif '(YPG) YA PAGO' in registro:
        return 17
    elif '(CNV) Convenio' in registro:
        return 18
    elif '(LTU) Liquidación Total Una Exhibición' in registro:
        return 19
    elif '(ALC) Credito al Corriente' in registro:
        return 20
    elif '(CRC) Cliente No Reconoce Credito' in registro:
        return 21
    else:
        return None  # Si no hay coincidencia

# Aplicar la función para crear la columna 'Categoria_Registro'
df['rgtro_cat'] = df['registro_final'].apply(asignar_registro_categoria)
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns

# Selección de variables numéricas relevantes
features_cluster = ['DIAS_VENCIDOS', 'SALDO_TOTAL', 'CanalCat', 'rgtro_cat', 'intentos_totales', 'dias_first','meses_first',
                          'horas_first']
X_cluster = df_nuevo[features_cluster]

# Estandarizar las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Determinar número óptimo de clusters con método del codo
sse = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(2, 10), sse, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('SSE (inercia)')
plt.title('Método del codo para elegir K')
plt.show()
```

```python
# Crear el modelo final de clustering
kmeans = KMeans(n_clusters=5, random_state=42)
df_nuevo['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Tasa de contacto exitoso por cluster
cluster_stats = df_nuevo.groupby('cluster')['Contacto'].agg(['mean', 'count'])
print(cluster_stats)

# Promedio de características por cluster
means = df_nuevo.groupby('cluster')[features_cluster].mean()
print(means)

# Visualización con seaborn
sns.barplot(data=df_nuevo, x='cluster', y='Contacto')
plt.title('Tasa de contacto exitoso por cluster')
plt.show()
```

```python
from kmodes.kmodes import KModes

#features_cluster = ['EDAD','dias_first','horas_first','CanalCat','Contacto']
features_cluster = ['dias_first','horas_first','CanalCat','Contacto']
df_discretizado = df_nuevo[features_cluster]
# 3. Convertir las columnas categóricas a un formato adecuado para K-Modes
# No necesitamos hacer una transformación adicional si las columnas ya son de tipo categórico

# 6. Mapear los valores numéricos de vuelta a sus categorías
dias_map = {
    1: 'Domingo',
    2: 'Lunes',
    3: 'Martes',
    4: 'Miércoles',
    5: 'Jueves',
    6: 'Viernes',
    7: 'Sábado'
}

def agrupar_hora(h):
    if 6 <= h < 11:
        return 0  # Mañana
    elif 11 <= h < 16:
        return 1  # Tarde
    elif 16 <= h <= 21:
        return 2  # Noche
    else:
        return 3  # Madrugada

df_discretizado.loc[:, 'horas_first'] = df_discretizado['horas_first'].apply(agrupar_hora)

horas_map = {
    0: 'Mañana',
    1: 'Tarde',
    2: 'Noche',
    3: 'Madrugada'
}

canal_map = {
    0: 'Sin canal',
    1: 'Llamada de Salida',
    2: 'WhatsApp',
    3: 'SMS',
    4: 'Llamada de Entrada',
    5: 'E-Mail',
    6: 'Judicial'
}

# Agregar segmentación de edad en rangos
def agrupar_edad(edad):
    if 20 <= edad <= 29:
        return '20-29'
    elif 30 <= edad <= 39:
        return '30-39'
    elif 40 <= edad <= 49:
        return '40-49'
    elif 50 <= edad <= 59:
        return '50-59'
    elif 60 <= edad <= 69:
        return '60-69'
    elif 70 <= edad <= 79:
        return '70-79'
    else:
        return '80+'

#df_contacto['RangoEdad'] = df_contacto['EDAD'].apply(agrupar_edad)
#df_discretizado.loc[:, 'EDAD'] = df_discretizado['EDAD'].apply(agrupar_edad)
# 1. Filtrar los datos donde hubo contacto exitoso (rgtro_cat = 'Éxito')
# Filtrar los valores válidos para 'dias_first' y 'horas_first'
df_discretizado = df_discretizado[df_discretizado['dias_first'].isin([1, 2, 3, 4, 5, 6, 7])]
df_discretizado = df_discretizado[df_discretizado['horas_first'].isin([0, 1, 2, 3])]
df_discretizado['CanalCat'] = df_discretizado['CanalCat'].map(canal_map)

# Aplicar el mapeo a las columnas categóricas
df_contacto = df_discretizado[df_discretizado['Contacto'] == 1]
df_contacto.head(5)
```

```python
# 2. Seleccionar las columnas relevantes: 'CanalCat', 'horas_first', 'dias_first'
# 4. Aplicar el algoritmo K-Modes
kmeans = KModes(n_clusters=5, init='Cao', n_init=5, verbose=1)
clusters = kmeans.fit_predict(df_contacto)

# 5. Agregar los clusters al DataFrame original para ver qué grupo pertenece cada registro
df_contacto['Cluster'] = clusters

# 7. Mostrar los centroides de los clusters (los valores más representativos de cada uno)
print("Centroides de los clusters:")
centroides = kmeans.cluster_centroids_

print("Valores de los centroides:")
print(centroides)
```
